# Sudoku

**What is it?**

Sudoku is a game that are played on a 9 by 9 grid where the player's objective is to place numbers 1 to 9 in the individual 3 by 3 arrays without repeating the same number in the same row or column. ALthough this game does require a large deal of numbers, it does not have any arithimetic required. More so, it requires logic and the processes of elimination to complete. There are usually different levels when it comes to this: easy levels typically have more numbers given on the grids where as harder ones are limited with numbers.
